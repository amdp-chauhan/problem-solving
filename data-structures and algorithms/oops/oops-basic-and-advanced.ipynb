# Class, Objects, and Access Modifiers

```python
class ExpenseTracker:
    """
    Class created just to learn oops concepts
    """
    # Class Attribute
    expense_tracker_version = 0.1
    
    # initialization function
    def __init__(self, track_category, original_balance, budget):
        # Instance/Object Attributes
        # Public - Accessible by the object
        self.tracking_category = track_category
        
        # Private Attribute - Accessible by the class only
        self.__original_balance = original_balance
        self.__tracker_budget = budget
        
        
    # instance method
    def __get_original_balance(self):
        return self.__original_balance
    
    # Instance Method
    def check_balance(self, limit=1000):
        if self.tracker_budget>limit:
            return True
        else:
            return "Your opening balance is less than limit"
    
    # Static Method
    @staticmethod
    def convert_amount(amount):
        return float(amount)

    # Factory Method
    @classmethod
    def get_attributes_from_string(cls, diary_entry: str):
        tracking_category, original_balance, tracker_budget = diary_entry.split(" ")
        
        return ExpenseTracker(tracking_category.capitalize(),
                             cls.convert_amount(original_balance),
                             cls.convert_amount(tracker_budget))
    
```

```python
# Child Class
class Car(Vehicle):
    def __init__(self, make, model, fuel, air_conditioning, sunroof):
        # Parent values
        Vehicle.make = make
        Vehicle.__model = model
        Vehicle.__fuel = fuel

        # Child values
        self.air_conditioning = air_conditioning
        self.sunroof = sunroof

    def show_parent_attributes(self):
        print("Make:",Vehicle.make, "Model:",Vehicle.__model, "Fuel:",Vehicle.__fuel)
        
    def show_privatemethod_ofparent(self):
        self._Vehicle__private_method_parent()
```

# Polymorphisim
Poly + Morphisim

Many + Forms - means, same method can be used for differenct purpose. For Eg - len(string)|len(list), method overriding

```python
# Parent Class
class Vehicle():
    current_year = 2021
    base_price = 1000

    def __init__(self, make, model, fuel):
        self.make = make
        # private
        self.__model = model
        self.__fuel = fuel

    def __private_method_parent(self):
        print("Accessed Private method.")

    def get_value(self):
        age = Vehicle.current_year - self.__model
        print("Default method of the parent.")
        return Vehicle.base_price*(1/age)

# Super function - They return the temporary instance of the class, protecting actual parent methods and objects

# Child Class
class CarWithSuper(Vehicle):
    def __init__(self, make, model, fuel, air_conditioning, sunroof):
        
        super(CarWithSuper, self).__init__(make, model, fuel)

        # Child values
        self.air_conditioning = air_conditioning
        self.sunroof = sunroof

    # Overriding Method - replaces the parent default method
    def get_value(self):
        Vehicle.base_price = 5000
        age = Vehicle.current_year - self._Vehicle__model
        print("This is child override method.")
        return Vehicle.base_price*(1/age)
```

## Protected

- attributes and methods starting with _. Note that it's not same a Private, where variable or method starts with __<br>
- Idea is to protect the parent&child class specific attributes, which are just needed for inner-class operations <br>



```python
# Parent Class
class Vehicle():
    current_year = 2021
    base_price = 1000

    def __init__(self, make, model, fuel):
        self.make = make
        # protected
        self._model = model
        self._fuel = fuel

    def __private_method_parent(self):
        print("Accessed Private method.")

    def _get_value(self):
        age = Vehicle.current_year - self.__model
        print("Default method of the parent.")
        return Vehicle.base_price*(1/age)

# Super function - They return the temporary instance of the class, protecting actual parent methods and objects

# Child Class
class CarWithSuper(Vehicle):
    def __init__(self, make, model, fuel, air_conditioning, sunroof):
        
        super(CarWithSuper, self).__init__(make, model, fuel)

        # Child values - protected
        self._air_conditioning = air_conditioning
        self._sunroof = sunroof

    # Overriding Method - replaces the parent default method
    # Protected - going to override a protected method
    def _get_value(self):
        Vehicle.base_price = 5000
        age = Vehicle.current_year - self._model
        print("This is child override method.")
        return Vehicle.base_price*(1/age)
```

```python
# Going to override a protected method
myobj = CarWithSuper("Tesla", 2019, "electric", True, True)
```

```python
myobj._air_conditioning, myobj._sunroof
```

```python
myobj._get_value()
```

## Multod Resolution Order - 

To keep the order defined for execution of methods

```python
class Car():
    def __init__(self, make, model, fuel):
        self.make = make
        self.model = model
        self.fuel = fuel
    
    def get_car_detail(self):
        return "The make of the car is", self.model," from car class "
    
class ElectricCar():
    def __init__(self, make, model, fuel):
        self.make = make
        self.model = model
        self.fuel = fuel
    
    def get_car_detail(self):
        return "The make of the car is", self.make," from Electric Car class "

# Multiple Inheritance
# Sequence of Inheritance i.e. Car then ElectricCar defines the precedence of execution in case of common method/attribute names.
class Taxi(Car, ElectricCar):
    def __init__(self, make, model, fuel):
        super().__init__(make, model, fuel)
```

```python
myobj = Taxi("Tesla", 2019, "Electric")
myobj.__dict__
```

```python
myobj.get_car_detail()
```

```python
# Method resolution order
Taxi.mro()
```

# Operator Overloading

We can overring commonly used functions by using the magic functions; we can change the default behaviour of built-in functions.


```python
class Vehicle:
    def __init__(self, make, model, fare):
        self.make = make
        self.model = model
        self.fare = fare

    # overloading the default print method of the object
    def __str__(self):
        return "The make of the car is {} and the model is of {} with a fare of {}".format(self.make, self.model, self.fare)

    # + operator override
    def __add__(self, other):
        return self.fare + other.fare
```

```python
myobj = Vehicle("Tesla", 2019, 40)
myobj2 = Vehicle("Ford", 2000, 80)
```

```python
print(myobj, '\n\n\n', myobj2)
```

```python
# We want the fare to be added when used with + operator
myobj+myobj2
# fails if __add_(self, other) is not defined
```

```python
from abc import ABC,abstractmethod

class A(ABC):

    @abstractmethod
    def fun1(self):
        print("function of class A called")

    @abstractmethod
    def fun2(self):
        pass

class B(A):
    def fun1(self):
        super().fun1()
    def fun2(self):
        print("function 2 called")
o = B()
o.fun1()
```

```python
class TextReaderAbstract(ABC):
    
    def __init__(self, path, filename):
        self.path = path
        self.filename = filename
        
    @abstractmethod # this is just an interface or "RULE"
    def get_path(self):
        pass
    
    @abstractmethod # this is just an interface or "RULE"
    def get_filename(self):
        pass
```

```python
class TextReader(TextReaderAbstract):
    
    # init is not compulsory to define
    # Abstract methods are compulsory, without these it'll give an error
    def get_path(self):
        return self.path

    # Abstract methods are compulsory, without these it'll give an error
    def get_filename(self):
        return self.filename
```

```python
class Vehicle():
    def __init__(self, make, model, fuel):
        self.make = make
        self.model = model
        self.fuel = fuel

    def get_value(self):
        try:
            age = 2021 - int(self.model)
            return 1000 * (1/age)
        except:
            try:
                age = 2021 - int(self.model)
                return 1000 * (1/age)
            except ZeroDivisionError:
                age = 2021 - int(self.model)
                return 1000 * (1)
            except:
                return "You have entered Model year in incorrect format"
```

```python
# Custom Exception
# Defining NegativeCarValue Class
class NegativeCarValue(Exception):
    def __init__(self, value, message="Car value cannot be negative"):
        self.value = value
        self.message = message
        super().__init__(self.message)
    def __str__(self):
        return f'{self.message} -> {self.value}'


# Example 
class Vehicle():
    def __init__(self, make, model, fuel):
        self.make = make
        self.model = model
        self.fuel = fuel
        self.current_year = 2021

    def get_value(self):
        age = self.current_year - self.model
        if age < 0:
            raise NegativeCarValue(age)
        else:
            return 1000*(1/age)
```

```python
myobj = Vehicle("Tesla", 2019, "electric")
myobj.get_value()
```
