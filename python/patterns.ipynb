### Print below pattern



```python
# Print pattern

print("*"*12)
for _ in range(3):
    for i in range(5):
        for j in range(4, i-1, -1):
            print("*", end="")
        for j in range(i+1):
            print(" ", end="")
        for j in range(i+1):
            print(" ", end="")
        for j in range(4, i-1, -1):
            print("*", end="")
        print(end='\n')
    for i in range(5):
        for j in range(i+1):
            print("*", end="")
        for j in range(4, i-1, -1):
            print(" ", end="")
        for j in range(4, i-1, -1):
            print(" ", end="")
        for j in range(i+1):
            print("*", end="")
        print(end="\n")
    print("*"*12)
```
