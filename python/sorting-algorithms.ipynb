## Bubble Sort 

```python
# Inplace Sorting 
# Bubble sort
def bubble(list):
    i = len(list)
    sorted=False
    while i>1 and not sorted:
        sorted = True
        for j in range(i-1):
            # print(list[j], list[j+1])
            if list[j]>list[j+1]:
                list[j], list[j+1] = list[j+1], list[j]
                sorted = False
        i -= 1
```

## Selection Sort

```python
def selection(list):
    i = 0
    while i<len(list):
        min = i
        for j in range(i+1, len(list)):
            # print(list[i], list[j])
            if list[min]>list[j]: min = j
        list[min], list[i] = list[i], list[min]
        i += 1
```

## Insertion Sort

```python
def insertion(list):
    i = 1
    while i<len(list):
        min = i
        for j in range(0, i):
            # print(j, i)
            if list[j]>list[i]:
                list[j], list[i] = list[i], list[j]
        i += 1
```

## Merge Sort

```python
# Merge Sort - Modify input arr 
def merge_sorted_list(l_list, r_list, arr):
    # merge sorted list
    l = 0 
    r = 0
    k = 0
    while l<len(l_list) and r<len(r_list):
        if l_list[l]<r_list[r]:
            arr[k] = l_list[l]
            l += 1
            k += 1
        else:
            arr[k] = r_list[r]
            r += 1
            k += 1

    while l<len(l_list):
        arr[k] = l_list[l]
        l += 1
        k += 1

    while r<len(r_list):
        arr[k] = r_list[r]
        r += 1
        k += 1

def mergeSort(arr):
    # Please add your code here
    if len(arr)<=1: 
        return arr
    mid = len(arr)//2
    left_list = arr[:mid]
    right_list = arr[mid:]

    mergeSort(left_list)
    mergeSort(right_list)

    merge_sorted_list(left_list, right_list, arr)

# Main
# n=int(input())
arr= list([2, 3, 0, 10, 9, 7, 5, 8, 90, 87]) # list(int(i) for i in input().strip().split(' '))
mergeSort(arr)
print(*arr)
```

## Quick Sort - 

Merge sort was not really doing much, it was just recursively dividing & merging the list, through function calls. On the other hand, quick sort, performs the first operation on it's own, that is, finding the first pivot element position, then performs the recursive calls to perform the similar operation on left and right partitions.


```python
def partition(arr, start, end):
    pivot_pos = start # init the point at the available start of array
    pivot_elem = arr[start]
    l = start
    while l<=end:
        if arr[l]<pivot_elem:
            pivot_pos += 1
        l += 1
    # print('pivot_pos',pivot_pos)
    arr[pivot_pos], arr[start] = arr[start], arr[pivot_pos]
    i = start
    j = end

    # bring items smaller than pivot to the left and larger than pivot to the right
    while i<j:
        if arr[i]<pivot_elem: # i is at the right position
            i += 1
        elif arr[j]>=pivot_elem: # j is at the right position
            j -= 1
        else: #Swapping needed 
            arr[i], arr[j] = arr[j], arr[i]
            i += 1
            j -= 1
    return pivot_pos

def quickSort(arr, start, end):
    # Please add your code here
    if start>=end:
        return

    pivot = partition(arr, start, end)
    quickSort(arr, start, pivot-1)
    quickSort(arr, pivot+1, end)

n=int(input())
arr=list(int(i) for i in input().strip().split(' '))
quickSort(arr, 0, n-1)
print(*arr)
```
